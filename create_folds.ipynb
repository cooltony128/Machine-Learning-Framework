```python
import pandas as pd
```

```python
from sklearn import model_selection
```

```python
df = pd.read_csv('TRAIN_CLEANED.csv')
```

```python
df.head()
```

```python
df['kfold'] = -1
```

```python
df = df.sample(frac=1).reset_index(drop=True) #we sample the whole thing
```

```python
kf = model_selection.StratifiedKFold(n_splits=5, shuffle=False, random_state=128)
```

```python
for fold, (trainindex ,validationindex) in enumerate(kf.split(X=df, y=df.label.values)):
    df.loc[validationindex, 'kfold']=fold
```

```python
df.to_csv('TRAIN_CLEANED_FOLDS.csv', index=False)
```
